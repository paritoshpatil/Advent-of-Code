```python
import re
file_path = 'inputs/3.txt'

with open(file_path, 'r') as file:
    file_content = file.read()

pattern = r'do\(\)|don\'t\(\)|mul\([0-9]{1,3},[0-9]{1,3}\)'
matches = re.findall(pattern, file_content)

print(matches)

final_result = 0
calculate_next = True
for match in matches:
    if match == 'don\'t()': 
        calculate_next = False
    elif match == 'do()':
        calculate_next = True
    else:
        if calculate_next:
            i = int(match[match.index('(') + 1 : match.index(',')])
            j = int(match[match.index(',') + 1 : match.index(')')])
            final_result += i * j

print(final_result)
```
